```python
import numpy as np
import pandas as pd
```

```python
credits=pd.read_csv('tmdb_5000_credits.csv')
movies=pd.read_csv('tmdb_5000_movies.csv')
```

```python
#merge both datasets
movies=movies.merge(credits,on="title")
```

```python
#columns needs to consider
#genres
#id
#keywords
#overview
#title
#cast
#crew
movies=movies[['id','title','overview','keywords','genres','cast','crew']]
```

```python
movies.dropna(inplace=True)
```

```python
import ast
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):   #convert obj that is in string to list so we can access it
        #through the library of ast and the function literal_eval
        L.append(i['name'])
    return L
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies.head(1)
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
#in cast data we want top 3 dictionaries of each movie
#in each dictionary there is a all info of one cast member(actor) so we want to fetch their name
```

```python
import ast
def convert3(obj):
    L=[]
    counter=0
    for i in ast.literal_eval(obj):   #convert obj that is in string to list so we can access it
        #through the library of ast and the function literal_eval
        if(counter!=3):
           L.append(i['name'])
           counter+=1
        else:
           break
    return L
```

```python
movies['cast']=movies['cast'].apply(convert3) 
```

```python
import ast
def fetch_director(obj):
    L=[]
    for i in ast.literal_eval(obj):   #convert obj that is in string to list so we can access it
        #through the library of ast and the function literal_eval
        if(i['job']=='director'):
          L.append(i['name'])
          break
    return L
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
#convert overview from string to list
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
#remove space between one entity ..so that system willnot confuse with same first named entity
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])   
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
new_df=movies[['id','title','tags']]
```

```python
new_df
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

## Text vectorization

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
cv.fit_transform(new_df['tags']).toarray().shape
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
features = cv.get_feature_names_out()
print(features[:100])  # First 100 words
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors) #calculating cosine distance between each vector-->less angle ->more similarity 
#cosine angle that is more with similar movies  
```

```python
cosine_similarity(vectors).shape  #as distance of each movie with every other movie
```

```python
#we need to hold the indexes in data so that after sorting ,if index is not there prediction may be wrong
#use enumerate ()
list(enumerate(similarity[0]))
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6] #we applied lambda bcoz from list we need sorting based on second column
#only top 5 similar movies
```

```python
def recommend(movie):
    #to fetch index of movie in data
    movie_index=new_df[new_df['title']==movie].index[0]
    distances=similarity[movie_index]
    movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    for i in movies_list:  #i is the whole tuple with index and cosine diatance of each of 5 movies
        print(new_df.iloc[i[0]].title)#printing by fetching through integer location of i[0]-->movie_index and column name is title
    
```
